```python
monthly_loan_set = ['monthly_bk_loan','monthly_cd_loan','monthly_installments_loan','monthly_insurance_loan','monthly_sbk_loan']
```

```python
ages_unique_list = list(data_set.ages.unique())
ages_unique_list
```

```python
loan_mean_list = list()
age_loan_dict = dict()
for age in list(data_set.ages.unique()):
    loan_mean = list()
    for loan in monthly_loan_set:
        loan_mean.append(data_set[data_set['ages']==age][loan].mean())
    loan_mean_list.append(loan_mean)

print(len(loan_mean_list))
loan_mean_list
```

```python
colors = sns.color_palette('coolwarm_r',len(loan_mean_list))

# plt.subplots_adjust(wspace=1)
plt.figure(figsize=(13,13))
plt.subplots_adjust(wspace=0.1)
title_font = {
    'fontsize': 40,
}

plt.pie(loan_mean_list[0],colors=colors)
plt.title(f'{(i+1)*10}\'s',fontdict=title_font)
plt.legend(labels=monthly_loan_set, loc='upper right',fontsize=10)

plt.show()
```

```python
from sklearn.model_selection import train_test_split
from xgboost import plot_importance
from xgboost import XGBClassifier
```

```python
y = data_set.avg_score
X = data_set.drop('avg_score',axis=1)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=1)
print(X_train.shape,X_test.shape)
```

```python
xgb = XGBClassifier(n_estimators=100,learning_rate=0.1,max_depth=4)
xgb.fit(X_train,y_train)
pred_train = xgb.predict(X_train)
pred_test = xgb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
def print_metrics(y, pred):
    acc = accuracy_score(y, pred)
    recall = recall_score(y, pred, average='micro')
    precision = precision_score(y, pred, average='micro')
    f1 = f1_score(y, pred, average='micro')
    print(f'정확도:{acc}, 재현율(recall):{recall}, 정밀도(precision):{precision}, f1:{f1}')
```

```python
print_metrics(y_test, pred_test)
```

```python
fig, ax = plt.subplots()
plot_importance(xgb,ax=ax,color='orangered')
plt.title('importance', fontsize = 15)
plt.yticks(fontsize = 10)
plt.ylabel('features', fontsize = 20)
```

```python
print_metrics(y_train, pred_train)
```
